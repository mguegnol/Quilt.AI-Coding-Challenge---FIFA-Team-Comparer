## Choosing interesting columns

```python
columns = ['Name', 
           'Age',
           'Photo',
           'Nationality',
           'Flag',
           'Overall', 
           'Potential', 
           'Club',
           'Club Logo',
           'Value', 
           'Wage', 
           'Special',
           'Preferred Foot', 
           'International Reputation', 
           'Weak Foot',
           'Skill Moves', 
           'Body Type', 
           'Position',
           'Jersey Number',
           'Height', 
           'Weight', 
           'Crossing', 'Finishing', 'HeadingAccuracy',
           'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy',
           'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed',
           'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping',
           'Stamina', 'Strength', 'LongShots', 'Aggression', 'Interceptions',
           'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking',
           'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
           'GKKicking', 'GKPositioning', 'GKReflexes']
```

## New DataFrame df

```python
df = pd.DataFrame(data, columns = columns)
```

```python
df.head()
```

## Checking null values

```python
df.isnull().sum()
```

## Filling null values

```python
df['Club'].fillna('No Club', inplace = True) # No club assigned
df['Position'].fillna('ST', inplace = True) # side midfielder position (moving player)
```

## Filling null values with mean for Discrete and Continuous variables

```python
to_impute_by_mean = df.loc[:, ['Crossing', 'Finishing', 'HeadingAccuracy',
                                 'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy',
                                 'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed',
                                 'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping',
                                 'Stamina', 'Strength', 'LongShots', 'Aggression', 'Interceptions',
                                 'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking',
                                 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
                                 'GKKicking', 'GKPositioning', 'GKReflexes']]
for i in to_impute_by_mean.columns:
    df[i].fillna(df[i].mean(), inplace = True)
```

## Filling null values with mode for categorical variables

```python
to_impute_by_mode = df.loc[:, ['Body Type','International Reputation', 'Height', 'Weight', 'Preferred Foot','Jersey Number']]
for i in to_impute_by_mode.columns:
    df[i].fillna(df[i].mode()[0], inplace = True)
```

## Filling null values with median for disrcrete numerical variables

```python
to_impute_by_median = df.loc[:, ['Weak Foot', 'Skill Moves', ]]
for i in to_impute_by_median.columns:
    df[i].fillna(df[i].median(), inplace = True)
```

## Checking for remaining null values

```python
df.isnull().sum().sum()
```

## Creating categories from skills

```python
def general(df):
    return int(round((df[['HeadingAccuracy', 'Dribbling', 'Curve', 
                            'BallControl']].mean()).mean()))

def mental(df):
    return int(round((df[['Aggression', 'Interceptions', 'Positioning', 
                            'Vision','Composure']].mean()).mean()))

def mobility(df):
    return int(round((df[['Acceleration', 'SprintSpeed', 
                            'Agility','Reactions']].mean()).mean()))

def power(df):
    return int(round((df[['Balance', 'Jumping', 'Stamina', 
                            'Strength']].mean()).mean()))

def shooting(df):
    return int(round((df[['Finishing', 'Volleys', 'FKAccuracy', 
                            'ShotPower','LongShots', 'Penalties']].mean()).mean()))

def passing(df):
    return int(round((df[['Crossing', 'ShortPassing', 
                            'LongPassing']].mean()).mean()))

def defending(df):
    return int(round((df[['Marking', 'StandingTackle', 
                            'SlidingTackle']].mean()).mean()))

def goalkeeping(df):
    return int(round((df[['GKDiving', 'GKHandling', 'GKKicking', 
                            'GKPositioning', 'GKReflexes']].mean()).mean()))

def rating(df):
    return int(round((df[['Potential', 'Overall']].mean()).mean()))
```

```python
df['General'] = df.apply(general, axis = 1)
df['Mental'] = df.apply(mental, axis =1)
df['Mobility'] = df.apply(mobility, axis = 1)
df['Power'] = df.apply(power, axis = 1)
df['Shooting'] = df.apply(shooting, axis = 1)
df['Passing'] = df.apply(passing, axis = 1)
df['Defending'] = df.apply(defending, axis =1)
df['Goalkeeping'] = df.apply(goalkeeping, axis = 1)
df['Rating'] = df.apply(rating, axis =1)
```

## Droping lazy features

```python
df.drop(columns = ['Crossing', 'Finishing', 'HeadingAccuracy',
                     'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy',
                     'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed',
                     'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping',
                     'Stamina', 'Strength', 'LongShots', 'Aggression', 'Interceptions',
                     'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking',
                     'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
                     'GKKicking', 'GKPositioning', 'GKReflexes', 'Overall', 'Potential'], inplace = True, axis =1)

df.columns.values
```

## Cleaning Height, Weight, Value and Wage

```python
'''For weight'''
def extract_value_from(value):
    x = value.replace('lbs', '')
    return float(x)

'''For Height'''
def feet_to_meters(value): 
    tmp = value.split("'")
    return (int(tmp[0]) + int(tmp[1]) * 0.1) * 0.3048 #converting feet to meters

'''For Value and wage'''
def convert_currency(value):
    x = value.replace('€', '')
    if 'M' in x:
        x = float(x.replace('M', ''))*100000
    elif 'K' in value:
        x = float(x.replace('K', ''))*1000
    return float(x)
```

```python
df['Weight'] = df['Weight'].apply(lambda x: extract_value_from(x))
df['Height'] = df['Height'].apply(lambda x: feet_to_meters(x))
df['Value'] =  df['Value'].apply(lambda x: convert_currency(x))
df['Wage'] =  df['Wage'].apply(lambda x: convert_currency(x))

df[['Weight', 'Height', 'Value', 'Wage']].head(3)
```
